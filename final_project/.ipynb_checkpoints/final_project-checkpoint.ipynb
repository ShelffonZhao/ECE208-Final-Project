```python
import numpy as np
import pandas as pd
```

# data cleaning

```python
df = pd.read_excel('trainset.xlsx')
```

```python
df.head()
```

```python
trainy_m = df.groupby("Compound Identifier").mean()
trainy_s = df.groupby("Compound Identifier").std()
trainy_m = trainy_m.drop(trainy_m.columns[0], axis=1)
trainy_s = trainy_s.drop(trainy_s.columns[0], axis=1)
```

```python
trainy_m.head()
```

```python
trainy_s.head()
```

```python
testset = pd.read_excel('testset.xlsx')
testy_m = testset.drop('sigma', axis = 1)
testy_s = testset.drop('value', axis = 1)
testy_m = testy_m.pivot(index='oID', columns = 'descriptor')
testy_m.columns = testy_m.columns.droplevel(0)
testy_s = testy_s.pivot(index='oID', columns = 'descriptor')
testy_s.columns = testy_s.columns.droplevel(0)
```

```python
testy_m.head()
```

```python
testy_s.head()
```

```python
molecular = pd.read_excel('molecular.xlsx',index_col='CID')
```

```python
molecular.head()
```

```python
trainx = molecular[molecular.index.isin(trainy_m.index.values)]
testx = molecular[molecular.index.isin(testy_m.index.values)]
```

```python
testx.head()
```

```python
trainx.head()
```

# variable

trainx: train x 

trainy_m : train y of mean

trainy_s : train y of std

testx : test x

testy_m : test y of mean

testy_s : test y of std
